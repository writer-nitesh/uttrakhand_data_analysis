# Uttrakhand Data Analysis

### Data used : Distict Wise India Census Analysis 2011


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("./dataset/census.csv", index_col="District_code")
data.head(2)
```

```python
ukdf = data[data["State_name"] == "uttarakhand".upper()]
ukdf.head(2)
```

```python
ukdf.shape, ukdf.size
```

# Population of Uttrakhand


```python
# Total Population of Uttrakhand
ukdf["Population"].sum()
```

```python
# Distict Wise Population of Uttrakhand in Ascending Order
ukdf[["District_name", "Population"]].sort_values(by="Population")
```

```python
plt.style.use("dark_background")
```

```python
plt.bar(x=ukdf["District_name"], height=ukdf["Population"])
plt.xlabel("District")
plt.ylabel("Population")
plt.xticks(rotation=90)
plt.title("Population by District")
plt.show()
```

# Distribution of people among different religions


```python
# Religion Wise Population of Uttrakhand in Ascending Order
religion = ['Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains']
ukdf[religion]
```

```python
plt.figure(figsize=(12, 6))
color = ['white', 'blue', 'purple', 'red', 'lightblue', "yellow"]

# Subplot 1: Bar Plot
plt.subplot(1, 2, 1)
plt.bar(x=religion, height=ukdf[religion].sum(), color=color)
plt.xlabel("Religions")
plt.ylabel("People")
plt.xticks(rotation=90)
plt.title("Distribution of people among different religions")

# Subplot 2: Pie Chart
plt.subplot(1, 2, 2)
plt.pie(ukdf[religion].sum(), labels=religion, colors=color)
plt.title("Percentage distribution of people among different religions")

plt.tight_layout()
plt.show()
```

# Total Number of Males and Females


```python
ukdf[["Male", "Female"]].sum()
```

```python
plt.pie(ukdf[["Male", "Female"]].sum(), labels=ukdf[["Male", "Female"]].sum())
plt.title("Gender")
plt.show()
```

# Distict wise Total Number of Males and Females


```python
district = ukdf["District_name"]
male = ukdf["Male"]
female = ukdf["Female"]
X_axis = np.arange(len(district))
```

```python
plt.figure(figsize=(10, 6))
plt.bar(x=X_axis-0.2, width=0.4, height=male, label='Male')
plt.bar(x=X_axis+0.2, width=0.4, height=female, label='Female')
plt.title("Distict wise Total Number of Males and Females")
plt.xticks(X_axis, district, rotation=90)
plt.xlabel("District")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
status = ukdf[['Secondary_Education',
               'Higher_Education', 'Graduate_Education']].sum()

lit = [ukdf["Population"].sum(), ukdf["Literate"].sum()]

color = ['red', 'lightblue', "green"]
labels = ["", "Literate"]
```

# Literacy Rate


```python
plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1)
plt.pie(lit, labels=labels, colors=["pink", "orange"], autopct='%1.1f%%')
plt.title("Literacy Rate")

plt.subplot(1, 2, 2)
plt.pie(status, colors=color, labels=[
        'Secondary Education', 'Higher Education', 'Graduate Education'], autopct='%1.1f%%')
plt.title("Qualification")
plt.show()
```

# Working Population

```python
total_man = ukdf["Male"].sum()
total_female = ukdf["Female"].sum()

working_men = ukdf["Male_Workers"].sum()
working_female = ukdf["Female_Workers"].sum()
work_labels = ["Male", "Female"]
```

```python
plt.figure(figsize=(10, 6))


plt.subplot(2, 3, 1)
plt.pie([total_man, total_female], labels=work_labels,
        colors=["pink", "orange"], autopct='%1.1f%%')
plt.title("Gender Working")

plt.subplot(2, 3, 2)
plt.pie([working_men, working_female], labels=work_labels,
        colors=['green', "blue"],  autopct='%1.1f%%')
plt.title("Working Gender")


plt.subplot(2, 3, 3)
plt.pie([total_man, working_men], labels=["Total Males", "Working Males"],
        colors=['lightgreen', "deepskyblue"],  autopct='%1.1f%%')
plt.title("Working Males")

plt.subplot(2, 3, 4)
plt.pie([total_female, working_female], labels=["Total Females", "Working Females"],
        colors=['fuchsia', "maroon"],  autopct='%1.1f%%')
plt.title("Working Females")

plt.subplot(2, 3, 5)
plt.pie([ukdf["Population"].sum(),(total_female+ working_female)], labels=["Non Working Poulation", "Working Population"],
        colors=['maroon', "fuchsia"],  autopct='%1.1f%%')
plt.title("Total Working Poulation")


plt.show()
```
